# step 2: install miniconda

Anaconda (1.3G) is too large to me, so I prefer using miniconda. 

Download from [here](https://conda.io/miniconda.html) (saved to ~/Download folder) and install

```
cd ~/Download
bash Miniconda3-latest-MacOSX-x86_64.sh -b -p $HOME/miniconda3
```

Since I am using zsh, so add PATH to ~/.zshrc:

```
vim ~/.zshrc
```

Find user configure and add as following
```
# User configuration
 export PATH="$HOME/miniconda3/bin:$PATH"
```

Save and exit (shift+zz)

# step 3: create a virtual environment and new kernel

Create virtual env named 'tfpy3':
```
conda create -n tfpy3 python=3
```

Activate virtual env 'tfpy3':
```
source activate tfpy3
```

Install new kernel (follow the instruction in the [document](https://ipython.readthedocs.io/en/latest/install/kernel_install.html#kernel-install):
```
conda install pip
conda install ipykernel # or pip install ipykernel

```

Install kernel into our virtual env 'tfpy3' and set kernel name as 'tfpy3':
```
python -m ipykernel install --user --name tfpy3  --display-name "tfpy3"
```

Make kernel in env available to other Jupyter (change xyza to your user name):
```
python3 -m ipykernel install --prefix='/Users/xyza/miniconda3/envs/tfpy3' --name 'tfpy3'
```
